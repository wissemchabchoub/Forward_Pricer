```python
def backtesting(N_j=30):
    results=[]
    N_j=int(ech1)
    for i in range(len(f)-N_j):
        d=str((f['date'][i+N_j]).date())
        if selected1.get()==1:
            fwd=forward(f['bid eur'][i+N_j],f['ask eur'][i+N_j],interpolation_lin_tmm(d),interpolation_lin_e(d)
                        ,interpolation_lin_tplc(d),interpolation_lin_eplc(d),ech1,mcc)
            spot=f['ask eur'][i]
            results.append(spot-fwd)  
        else:
            fwd=forward(f['bid usd'][i+N_j],f['ask usd'][i+N_j],interpolation_lin_tmm(d),interpolation_lin_u(d)
                        ,interpolation_lin_tplc(d),interpolation_lin_usdplc(d),ech1,mcc)
            spot=f['ask usd'][i]
            results.append(spot-fwd)
            
    plt.switch_backend('TkAgg')
    
    txt='Espérance = '+"%.4f" % np.mean(results) + '     Variance = '+("%.4f" % np.std(results))
    plt.text(0.2,0.2,txt)
    
    plt.plot(results)
    plt.title('Backtesting')
    plt.xlabel('Historique')
    plt.ylabel('Loss and Gain')
    
    
    return results
```
